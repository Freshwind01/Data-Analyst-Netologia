```python
#3
df3=pd.read_csv("ratings.csv",sep=',')
```

```python
rate=df3.groupby(['userId']).count()['rating'].reset_index()  #группирую по юзеру, оставляю столбец рейтинг
```

```python
rate1=rate[rate['rating']>100].sort_values(['userId'])     #Оставляем только больше 100. 
```

```python
final_data=df3[df3['userId'].isin(rate1['userId'])][['userId','timestamp']]     #из основной таблицы выбираю только те строчки, которые есть в выборке
final_data
```

```python
rating_max=final_data.groupby('userId').max()['timestamp'].reset_index()
rating_min=final_data.groupby('userId').min()['timestamp'].reset_index()
lifetime=rating_max['timestamp']-rating_min['timestamp']
res =pd.concat([rating_max['userId'],lifetime],axis=1)
res
```
