Успешных комбинаций 9.
Всего 3 барабана по 10 цифр на каждом. Количество возможных комбинаций 10 в кубе.
Количество вариантов положительного исхода (выигрыш):
*   5 комбинаций из одинаковых нечетных чисел;
*   По 10 комбинаций типа *77 и *00 (20)
*   По 100 комбинаций типа * * 7 и * * 0  (200)
Итого 225 устраивающих нас исходов.
Теоретическая вероятность **0.225**





 


```python
def get_win(A,B,C):   #Функция для подсчета выигрыша
  summa=0
  if (A==B==C):
      if A==7:
          summa=200
      elif A==9:
          summa=100
      elif A==5:
          summa=50
      elif A==3:
          summa=15 
      elif A==1:
          summa=10
  else:    
      if    B==C==7:
              summa=5
      elif  B==C==0:
              summa=2
      elif C==7:
              summa=3
      elif C==0:
              summa=1             
  return summa      
```

```python
trials = 1000000   # Количество попыток
wins = 0    # Количество успехов
s=[]        #Список с выигрышами, можно было просто складывать выигрыши.

for i in range(trials):
   A = np.random.randint(0, 10)
   B = np.random.randint(0, 10) 
   C = np.random.randint(0, 10)     # Случайное число от 0 до 9
   ABC=int(str(A)+str(B)+str(C))   # Собираю трехзначное число. Str, чтобы ноль не терялся.
   BC=int(str(B)+str(C))           # Двузначное число, чтобы проверять 77 и 00
   if (A!=0)&(A%2!=0):             # Если первая цифра не 0 и не четное, то...
     if A==B==C:
        wins += 1
        s.append(get_win(A,B,C))
#        print(ABC)                 #Это для отладки
     elif (ABC%10==7 or ABC%10==0) :    #По совету использовала деление с остатком
        wins += 1
        s.append(get_win(A,B,C))
#        print(ABC) 
   else:  
        if (BC%10==7 or BC%10==0 or BC==0) :      #При преобразовании BC второй ноль потеряется, поэтому должно так сработать для 00.     
           wins += 1 
           s.append(get_win(A,B,C))
#           print(str(A)+str(B)+str(C)) 
     
p_monte_carlo = wins / trials    # Вероятность из моделирования
p = 225 / 1000                            # Теоретическая вероятность
#print(s)   #Зачем-то я придумала список, что-то считать хотела по нему. А он вроде и не нужен оказался.
print('Выигрышей',wins,'Попыток', trials, p_monte_carlo, p,)
print('За серию из',trials,'игр сумма выигрыша составила:',sum(s),'руб. Потрачено:',trials,'рублей.')
print('Средний выигрыш без учета потраченных на игру денег:',round(sum(s)/trials,2))
print('Средний доход игрока за одну игру составляет', round((sum(s)-trials)/trials,2))
```

А вот при неоднократном повторении получается, что все зависит от количества 777. Но вероятность выпадения 7-ки ничем не отличается от вероятности выпадения 1. Вот это мне непонятно.

Но может и это тоже как-то можно заложить, тогда вероятность еще ниже. 

**P=0.1 * 0.1 * 0.1= 0.001**

Выигрыши могут быть 200,100,50,15,10.

2. Две одинаковые цифры:

**P=0.9 * 0.1 * 0.1=0.009**

0.9 потому что первая цифра не может совпадать с двумя другими и выбирается, соответственно, из 9-ти оставшихся.

Выигрыши могут быть 5 и 2

3. В конце 0 или 7. 
**P=10/10 * 0.9 * 0.1=0.09**

Первая цифра любая - 10 из 10, вторая не может совпадать с третьей.
Выигрыши могут быть 1 и 3
